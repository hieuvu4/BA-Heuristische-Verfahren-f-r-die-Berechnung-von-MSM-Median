```python
names = ["50words", "Adiac", "Beef", "CBF", "Coffee", "ECG200", "FaceAll", "FaceFour", "FISH", "Gun_Point", "Lighting2", "Lighting7", "OliveOil", "OSULeaf", "SwedishLeaf", "synthetic_control", "Trace", "Two_Patterns", "wafer", "yoga"]
k = 3
n = 18
classType = "mixedClasses"
# classType = "oneClass"
```

```python
for name in names:
    # path of samples
    file = open("/Users/hieu/Library/CloudStorage/OneDrive-Personal/Studium/BA Heuristische Verfahren fr die Berechnung von MSM Median/randomSamplesUCRArchive/k_" + str(k) + "/n_" + str(n) + "/" + classType + "/" + name + "_TRAIN_k_" + str(k) + "_n_" + str(n) + ".csv")
    data = np.loadtxt(file, delimiter=",")
    # print(data)
    plt.title(name)
    plt.xlabel("n")
    plt.gca().get_yaxis().set_visible(False)
    xlabels = [i+1 for i in range(n)]
    for i in range(len(data)):
        plt.plot(xlabels, data[i], label="ts" + str(i+1), alpha = 0.5)
    plt.legend(loc="upper left")
    plt.show()
```
